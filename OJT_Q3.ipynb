**SVD** **Decomposition**

```python
import numpy as np
import sympy as sp
np.random.seed(6)
A = np.random.randint(0,10,(5,5))
print(A)
```

```python
sp.Matrix(A)
```

```python
U,S,Vt=np.linalg.svd(A)
sp.Matrix(np.round(U))
```

```python
sp.Matrix(U)
```

```python
sp.Matrix(Vt)
```

```python
Sigma=np.zeros((A.shape[0],A.shape[1]))
```

```python
Sigma[:A.shape[1],:A.shape[0]]=np.diag(S)
sp.Matrix(Sigma)
```

```python
U1=U[:,:2]
U1=sp.Matrix(U1)
U1
```

```python
Vt1=Vt[:2,:]
Vt1=sp.Matrix(Vt1)
Vt1
```

```python
Sig1=Sigma[:2,:2]
Sig1=sp.Matrix(Sig1)
Sig1
```

```python
A2=U1@Sig1@Vt1
print('Rank 2 Aproximation of A')
A2
```

```python
U2=U[:,:3]
U2=sp.Matrix(U1)
U2
```

```python
Vt2=Vt[:3,:]
Vt2=sp.Matrix(Vt2)
Vt2
```

```python
Sig2=Sigma[:3,:3]
Sig2=sp.Matrix(Sig2)
Sig2
```

```python
A3=U1@Sig1@Vt1
print('Rank 3 Aproximation of A')
A3
```
